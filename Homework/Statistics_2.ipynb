### Задание 5

```python
def setosa_look(x):
  if (x['petal.length'] < 2.0) & (x['petal.width'] < 1.7):
    return 'Setosa'
  else: 
    return 'Unknown'
```

### Задание 6

```python
df['setosa_type2']  = df.apply(setosa_look, axis=1)
```
